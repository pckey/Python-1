```python
import requests
from  bs4 import BeautifulSoup
import os
import lxml
import numpy as np

#获取url下的页面内容，返回soup对象
def get_page(url):
    responce = requests.get(url)
    soup = BeautifulSoup(responce.text,'lxml')
    return soup
```

```python
#获取网页数据
urlcount=1000
url ="http://chart.icaile.com/chart/yn/?top="+str(urlcount)
soup = get_page(url)
```
