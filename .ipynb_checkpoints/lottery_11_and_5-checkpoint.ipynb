```python
# -*- coding:utf-8 -*-

import requests
from  bs4 import BeautifulSoup
import os
import lxml
#获取url下的页面内容，返回soup对象
def get_page(url):
    responce = requests.get(url)
    soup = BeautifulSoup(responce.text,'lxml')
    return soup


#封装成函数，作用获取列表页面下的所有信息，返回一个连接列表
def get_links(link_url):
    soup = get_page(link_url)
    links_li = soup.find_all('li')
    links = [div.a.get('href') for div in links_li]
    return links


```
